```python
# File Handling
# 1️⃣ Open a File

file = open("data.txt","r")
print(file.read())
file.close()
```

```python
# Best Practice – Using "with"

with open("data.txt", "r") as file:
    content = file.read()
    print(content)
```

```python
#Write to File
# "w" Overwrite the data

with open("data.txt", "w") as file:
    file.write("Pune, Maharashtra")
    file.write("Tanaji")
   
```

```python
# Append to File
#Adds at end.

with open("data.txt", "a") as file: 
    file.write("Tanaji More")
```

```python
# Read Line by Line

with open("data.txt", "r") as file:
    for line in file:
        print(line.strip())
```

```python
#Create a file and write numbers 1 to 5 into it.

with open("new_Data.txt", "w") as file:
    file.write("1,2,3,4,5")
```

```python
#3️⃣ Read that file and print content.'

with open ("new_Data.txt", "r") as file:
    print(file.read())
```

```python
with open("numbers.txt", "w") as file:
    for i in range(1,10):
        file.write(str(i)+ "\n")
```

```python
with open("numbers.txt", "r") as file:
    con = file.read()
    print(con)
```
