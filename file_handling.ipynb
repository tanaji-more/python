```python
# File Handling
# 1️⃣ Open a File

file = open("data.txt","r")
print(file.read())
file.close()
```

```python
# Best Practice – Using "with"

with open("data.txt", "r") as file:
    content = file.read()
    print(content)
```

```python
#Write to File
# "w" Overwrite the data

with open("data.txt", "w") as file:
    file.write("Pune, Maharashtra")
    file.write("Tanaji")
   
```

```python
# Append to File

with open("data.txt", "a") as file: 
    file.write("Tanaji More")
```
