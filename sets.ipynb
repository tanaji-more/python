```python
a = {1,2,3,4,5}
b = {5,6,7,8,1}

print(a.union(b)) # all items from both sets, duplicates will be removed
print(a.intersection(b)) # items that are present in both sets
print(a.difference()) # items that are present in set a but not in set b
print(a|b)
print(a.symmetric_difference(b))
```
