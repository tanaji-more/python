```python
#Given a list of daily sales: sales = [1200, 1500, 1100, 1800, 1700, 1600]
# Find total revenue."""

sales = [1200, 1500, 1100, 1800, 1700, 1600]

total_revenue= sum(sales)
print("Total_revenue", total_revenue)
```

```python
# 3. Remove all sales less than 1300.

sales = [1200, 1500, 1100, 1800, 1700, 1600]

filtered_sales = [sale for sale in sales if sale >= 1300]
print(filtered_sales)
```
