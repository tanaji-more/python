```python
# 1Ô∏è‚É£ Frequency Count (Very Important)

cities = ["Pune", "Mumbai", "Pune", "Delhi", "Mumbai", "Pune"]

city_count = {}

for city in cities:
    if city in city_count:
        print(city_count)
        city_count[city] += 1
    else: 
        city_count[city] = 1

print(city_count)
```

```python
# 3Ô∏è‚É£ Total Annual Sales
sales = {
    "Jan": 50000,
    "Feb": 65000,
    "Mar": 48000,
    "Apr": 70000
}

total_sales= sum(sales.values())
print(total_sales)
```

```python
# 4Ô∏è‚É£ Sort Dictionary by Values (Descending)

sales = {
    "Jan": 50000,
    "Feb": 65000,
    "Mar": 48000,
    "Apr": 70000
}

sorted_sales = dict(sorted(sales.items(), key= lambda x:x[1], reverse = True))
print(sorted_sales)

```

```python
#5Ô∏è‚É£ Merge Two Dictionaries (Add Duplicate Values)

dept1 = {"HR": 5, "IT": 10}
dept2 = {"Fianance": 4, "IT": 3}

merged = dept1.copy()

for key, value in dept2.items():
    if key in merged:
        merged[key] += value
    else:
        merged[key] = value

print(merged)
```

```python
# 7Ô∏è‚É£ Dictionary Comprehension 

square = {num : num ** 2 for num in range(1,6)}
print(square) 
```

```python
# 8Ô∏è‚É£ Remove Students with Marks < 50
marks = {"Rahul": 33, "Amit": 78, "Priya": 44, "Sneha": 30 }

filtered_marks = {name: score for name, score in marks.items() if score>=50}

print(filtered_marks)


```
