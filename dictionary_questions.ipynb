```python
# 8Ô∏è‚É£ Find employee with highest salary
emp= {
    "Tanaji": 75000,
    "Rahul": 45000,
    "Sneha": 43000,
    "Sidhu": 87000
}

print("Higest Salary employee is", max(emp.keys()) ," and salary is ", max(emp.values()))

high_sal = max(emp, key= emp.get)
print(high_sal)
print(emp[high_sal])
```

```python
# 1Ô∏è‚É£ Frequency Count (Very Important)

cities = ["Pune", "Mumbai", "Pune", "Delhi", "Mumbai", "Pune"]

city_count = {}

for city in cities:
    if city in city_count:
        print(city_count)
        city_count[city] += 1
    else: 
        city_count[city] = 1

print(city_count)
```
