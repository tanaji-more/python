```python
#6️⃣ Print all employee names
emp= {
    "Tanaji": 75000,
    "Rahul": 45000,
    "Sneha": 43000,
    "Sidhu": 87000
}

print(emp.keys())
print(list(emp.keys()))
print(emp.values())

```

```python
# 7️⃣ Print all salaries

emp= {
    "Tanaji": 75000,
    "Rahul": 45000,
    "Sneha": 43000,
    "Sidhu": 87000
}

print("All salaries : ", list(emp.values()))
```

```python
# 8️⃣ Find employee with highest salary
emp= {
    "Tanaji": 75000,
    "Rahul": 45000,
    "Sneha": 43000,
    "Sidhu": 87000
}

print("Higest Salary employee is", max(emp.keys()) ," and salary is ", max(emp.values()))

high_sal = max(emp, key= emp.get)
print(high_sal)
print(emp[high_sal])
```
