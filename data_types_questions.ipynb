```python
# 1.Create a new list containing only numbers greater than 30.
numbers = [10, 20, 30, 40, 50, 60]

new_list = [i for i in numbers if i >30]
print(new_list)
```

```python
# 2. Find the second largest number in a list.

numbers = [10, 20, 30, 40, 50, 60, 60]

unique_no = list(set(numbers))
unique_no.sort()
second_higest = unique_no[-2]
print(second_higest)
```
