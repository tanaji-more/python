```python
# 1.Create a new list containing only numbers greater than 30.
numbers = [10, 20, 30, 40, 50, 60]

new_list = [i for i in numbers if i >30]
print(new_list)
```

```python
# 2. Find the second largest number in a list.

numbers = [10, 20, 30, 40, 50, 60, 60]

unique_no = list(set(numbers))
unique_no.sort()
second_higest = unique_no[-2]
print(second_higest)
```

```python
#3Ô∏è‚É£Remove duplicates from - (Without converting directly to set)

data = [1,2,3,2,4,5,1,6]

# method 1

unique_value= list(set(data))
print(unique_value)

# method 2
data = [1,2,3,2,4,5,1,6]
unique_list = []

for num in data:
    if num not in unique_list:
        unique_list.append(num)
print(unique_list)

# ‚úÖ Method 2: Using List Comprehension

data = [1,2,3,2,4,5,1,6]

unique_list= []
[unique_list.append(x) for x in data if x not in unique_list]
print(unique_list)

# Using Dictionary (Python 3.7+ Order Preserved

data = [1,2,3,2,4,5,1,6]

unique_list = list(dict.fromkeys(data))
print(unique_list)
```

```python
# 4Ô∏è‚É£Find common elements between two lists.

list1= [1,2,3,4,5,6,6,7]
list2= [3,4,5,6,7,7,8,9]

common = []
for i in list1:
    if i in list2:
        common.append(i)

print(common)
```

```python
# 5Ô∏è‚É£Flatten this nested list:

nested = [[1,2], [3,4], [5,6]]

# üîπ Method 1: Using Loop (Best for Interview)

flat_list= []

for sublist in nested:
    for item in sublist:
        flat_list.append(item)

print(flat_list)
```
