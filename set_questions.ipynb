```python
# 1.Find common customers. 

customers_day1 = {"A", "B", "C", "D"}
customers_day2 = {"C", "D", "E", "F"}

print(customers_day1.intersection(customers_day2))
print(customers_day1 & customers_day2)
```

```python
# 2. Find customers who visited only on day1.

customers_day1 = {"A", "B", "C", "D"}
customers_day2 = {"C", "D", "E", "F"}
print(customers_day1.difference(customers_day2))
print(customers_day1 - customers_day2)
```

```python
# 3. Add a new customer "G".
customers = {"A", "B", "C", "D"}
customers.add("G")
print(customers)
customers.update("TM")
print(customers)
```

```python
# 4. Remove customer "B".
customers = {"A", "B", "C", "D"}
customers.remove("B") # remove() gives error if value not found.
print(customers)
customers.discard("T") # If value may not exist â†’ use discard() instead.
print(customers)
```

```python
# 5. Find total unique customers from both days.

days1 = {"Tanaji", "More", "Sidhu"}
days2 = {"sidhu", "Tanaji", "Sidhu", "Tana"}

print(days1.union(days2))
print(days1 | days2)

```

```python
# 8. Perform union of two sets.

cust1 = {"Tanaji", "More", "Sidhu"}
cust2 = {"sidhu", "Tanaji", "Sidhu", "Tana"}

print(cust1.union(cust2))
```

```python
# 9. Perform difference of two sets.

cust1 = {"Tanaji", "More", "Sidhu"}
cust2 = {"sidhu", "Tanaji", "Sidhu", "Tana"}
print(cust1.difference(cust2))
print(cust1.symmetric_difference(cust2))
```
