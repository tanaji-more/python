```python
# this is normal
try:
    num = int(input("Enter the number"))
    result = 100 / num
    print(result)
except:
    print("Something Went wrong")
```

```python
# Specific exceptions

try:
    num= int(input("Enter the number : "))
    result= 100/ num 
    print("Result: ", result)
except ZeroDivisionError:
    print("Cannot devide by zero")

except ValueError:
    print("Please enter the a valid number")
```

```python
# üîπ finally Block

try:
    print(10/2)
except:
    print("Error")
finally:
    print("Done")
```

```python
# 1Ô∏è‚É£Create a file numbers.txt
# Write numbers 1‚Äì10.

with open("numbers.txt", "w") as file:
    for i in range(1,11):
        file.write(str(i) + "\n")
```

```python
# Read that file. Calculate sum of all numbers.

total = 0

with open("numbers.txt", "r") as file:
    for line in file:
        number = int(line.strip()) # it removes the new line (\n) and convert it into string
        total = total + number

print("Total sum : ", total)

with open('numbers.txt', 'r') as file:
    total = sum(int(line.strip()) for line in file)
print("Total sum : ", total)
 
```
