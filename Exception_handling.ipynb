```python
# this is normal
try:
    num = int(input("Enter the number"))
    result = 100 / num
    print(result)
except:
    print("Something Went wrong")
```

```python
# Specific exceptions

try:
    num= int(input("Enter the number : "))
    result= 100/ num 
    print("Result: ", result)
except ZeroDivisionError:
    print("Cannot devide by zero")

except ValueError:
    print("Please enter the a valid number")
```

```python
# üîπ Catching Error Message

try: 
    x = int("abc")
except Exception as e:
    print("Error  is : ", e)
```

```python
# üîπ finally Block

try:
    print(10/2)
except:
    print("Error")
finally:
    print("Done")
```
