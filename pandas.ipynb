```python
# Handling Missing values
import pandas as pd
data= {
    "Name": ["Tanaji", "Rahul", "Amit", "Priya", "Neha"],
    "Age": [25, 30, None, 27, 29],
    "Salary": [50000, 60000, 55000, None, 52000],
    "City": ["Pune", "Mumbai", "Delhi", "Pune", None]
}

df = pd.DataFrame(data)
print(df)
```

```python
# Count missing values
df.isnull().sum()
```

```python
df["City"] = df["City"].fillna("Unkown")
print(df)
```

```python
df['Salary'] = df['Salary'].fillna(0)
print(df)
```

```python
import pandas as pd

data= {
    "Name": ["Tanaji", "Rahul", "Amit", "Priya", "Neha", "Arjun"],
    "City": ["Pune", "Mumbai", "Pune", "Delhi", "Mumbai", "Pune"],
    "Department": ["IT", "HR", "IT", "Finance", "HR", "IT"],
    "Salary": [50000, 60000, 55000, 65000, 62000, 58000],
    "Age": [25, 30, 28, 32, 29, 27]
}

df = pd.DataFrame(data)
print(df)
```

```python
# 1. unique() ‚Üí Find Unique Values

df["City"].unique()

```

```python
# count unique values

df["City"].nunique()
```

```python
# 2. value_counts() ‚Üí Count Frequency

df['City'].value_counts()
```

```python
df["Department"].value_counts()
```

```python
# 3. Filtering with Multiple Conditions
# Example 1: Salary > 55000 AND City = Pune

df[(df["Salary"] > 55000) & (df["City"] == "Mumbai") | (df["City"] == "Pune")]
```

```python
# 2: Salary > 60000 OR Department = HR

df[(df["Salary"] > 60000) | (df["Department"] == "HR")]
```

```python
# Age between 25 and 30

df[(df["Age"] >= 25) & (df["Age"] <= 30)]
```

```python
# groupby() ‚Üí MOST IMPORTANT FUNCTION

# 1. Average salary per city


df.groupby("City")["Salary"].mean()
```

```python
# 2 Total salary per department

df.groupby("Department")["Salary"].sum()
```

```python
# 3: Count employees per department
df.groupby("Department")["Name"].count()
```

```python
# 4: Multiple aggregations

df.groupby("City")["Salary"].agg(["mean", 'max', 'min'])
```
