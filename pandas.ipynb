```python
# Handling Missing values

data= {
    "Name": ["Tanaji", "Rahul", "Amit", "Priya", "Neha"],
    "Age": [25, 30, None, 27, 29],
    "Salary": [50000, 60000, 55000, None, 52000],
    "City": ["Pune", "Mumbai", "Delhi", "Pune", None]
}

df = pd.DataFrame(data)
print(df)
```

```python
# üü¢ 1. Handling Missing Values

df.isnull()
```

```python
# Count missing values
df.isnull().sum()
```

```python
# Remove rows with missing values

df_clean = df.dropna()
print(df_clean)
```

```python
df["Age"] = df["Age"].fillna(0)
print(df)
```

```python
df["City"] = df["City"].fillna("Unkown")
print(df)
```
